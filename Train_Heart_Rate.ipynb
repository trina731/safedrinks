```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scientific Reference: https://link.springer.com/content/pdf/10.1007%2F978-3-319-29175-8.pdf
data_points = 1000

# Generate heart rate data from 60 - 100
heart_rate = np.linspace(60, 100, data_points)
# Generate BAC data from 0 - 0.8
bac = np.linspace(0, 0.8, data_points)

# Visualize heart rate data
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(heart_rate)
plt.xlabel("Time")
plt.ylabel("Heart Rate")

plt.subplot(1, 2, 2)
plt.plot(bac)
plt.xlabel("Time")
plt.ylabel("Blood Alcohol Content (BAC)")

plt.tight_layout()

# Rescale heart rate using max and min
min_hr = heart_rate.min()
heart_rate = heart_rate - min_hr
max_hr = heart_rate.max()
heart_rate = heart_rate / max_hr
heart_rate = np.reshape(heart_rate, (data_points, 1))
```

```python
# Train SVM for going from heart rate to BAC
from sklearn import svm
clf = svm.SVR()
clf.fit(heart_rate, bac)
```

```python
# Use SVM to predict BAC
clf.predict([[0.5]])
```
